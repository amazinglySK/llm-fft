## Interpretation of the metrics

### SNR
A high positive SNR (e.g., > 10 dB): The signal you kept is much stronger than the part you removed. This implies your filter made a "subtle" change.
An SNR around 0 dB: The kept signal and the removed component have roughly the same power.
A negative SNR (e.g., < 0 dB): The component you removed was actually stronger than the part you kept. This implies your filter made a very "aggressive" change.

### NRMSE (by stddev)
An NRMSE less than 1.0 indicates that your model's prediction error is smaller than the natural variability of the data, suggesting your model is better than simply predicting the mean.
An NRMSE greater than 1.0 suggests your model is performing worse than a simple mean-based prediction.
A value close to 1.0 indicates your model's performance is on par with the data's inherent variability.

### Percentage energy loss
A value of x% means that your filter stripped away components that constituted x% of the original signal's total energy. This is a great, high-level number for reports and presentations.
